```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('user-walking-activity/2.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
dados_X = df['x_acceleration'].to_numpy()
dados_Y = df['y_acceleration'].to_numpy()
dados_Z = df['z_acceleration'].to_numpy()
timestep = df['time_step'].to_numpy()

# Cria o gráfico 3D
fig = plt.figure(figsize=(15, 15))

ax = plt.axes(projection='3d')

# Adiciona os dados ao gráfico
ax.plot3D(timestep,dados_X, dados_Y, dados_Z,'red')
ax.view_init(elev=200, azim=89)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Gráfico de trajetória')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Dados de exemplo
dados_x = df['x_acceleration'].to_numpy()
dados_y = df['y_acceleration'].to_numpy()
dados_z = df['z_acceleration'].to_numpy()
timestep = df['time_step'].to_numpy()


fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(timestep, dados_x, dados_y, dados_z, c=timestep, cmap='rainbow')
ax.view_init(elev=200, azim=89)
ax.set_xlabel("Tempo (s)")
ax.set_ylabel("Aceleração em X (m/s^2)")
ax.set_zlabel("Aceleração em Y (m/s^2)")
plt.colorbar(label="Tempo (s)")
plt.show()
```
