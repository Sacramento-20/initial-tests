```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import do dataset
df = pd.read_csv("pedestrian/A_2017.11.21_15.03.50.csv")
```

```python
df.head()
```

```python
df
```

```python
# Descrição dos dados no dataset
df.describe()
```

```python
# DataFrame apenas com as colunas necessarias para a plotagem de uma trajetoria 
trajeto = df.loc[:, ["Timestamp", "Latitude", "Longitude"]]
trajeto
```

```python
# Trajeto do usuário
trajeto.plot.scatter(y='Latitude', x= 'Longitude', c='Latitude', cmap='cool',figsize= (10,5))
plt.grid()
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Trajeto 1')
plt.show()
```
