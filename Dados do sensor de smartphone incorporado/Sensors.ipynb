```python
# Trajeto Latitude e Longitude
```

```python
trajeto_coordenadas = df.loc[:, ["Date_time", "SENSORGPS_LATITUDE", "SENSORGPS_LONGITUDE"]]
trajeto_coordenadas
```

```python
trajeto_coordenadas.describe()
# Alguns valores de latitude e longitude possuem valores iguais a 0, para isso foi feita uma filtragem para que apenas sejam pegos os valores validos do trajeto.
```

```python
Trajeto_coordenadas = trajeto_coordenadas.loc[(trajeto_coordenadas["SENSORGPS_LATITUDE"] > 23) & (trajeto_coordenadas["SENSORGPS_LONGITUDE"] > 23)]
```

```python
Trajeto_coordenadas
```

```python
Trajeto_coordenadas.describe()
```

```python
# Trajetos do usuÃ¡rio
trajeto_coordenadas.plot.scatter(y='SENSORGPS_LATITUDE', x= 'SENSORGPS_LONGITUDE', c='SENSORGPS_LATITUDE', cmap='cool',figsize= (10,5))
plt.grid()
plt.xlabel('SENSORGPS_LONGITUDE')
plt.ylabel('SENSORGPS_LATITUDE')
plt.title('Trajeto')
plt.show()
```
